```python
import sys
sys.path = ['/Users/sebastian/github/mlxtend'] + sys.path

import mlxtend
from mlxtend.data import iris_data
from mlxtend.evaluate import plot_decision_regions
from mlxtend.classifier import Perceptron
mlxtend.__version__
```

## Loading Iris Data

```python
X, y = iris_data()
X = X[:, [0, 3]] # sepal length and petal width
X = X[0:100] # class 0 and class 1
y = y[0:100] # class 0 and class 1

# standardize
X[:,0] = (X[:,0] - X[:,0].mean()) / X[:,0].std()
X[:,1] = (X[:,1] - X[:,1].mean()) / X[:,1].std()
```

## Rosenblatt Perceptron Rule

```python
%matplotlib inline
import matplotlib.pyplot as plt


ppn = Perceptron(epochs=15, eta=0.01, learning='perceptron')
ppn.fit(X, y)

plot_decision_regions(X, y, clf=ppn)
plt.title('Perceptron - Rosenblatt Perceptron Rule')
plt.show()

print(ppn.w_)

plt.plot(range(len(ppn.cost_)), ppn.cost_)
plt.xlabel('Iterations')
plt.ylabel('Missclassifications')
plt.show()
```

## Gradient Descent Perceptron (Widrow Rule)

```python
%matplotlib inline
import matplotlib.pyplot as plt


ppn = Perceptron(epochs=15, eta=0.01, learning='gd')
ppn.fit(X, y)

plot_decision_regions(X, y, clf=ppn, )
plt.title('Perceptron - Gradient Descent')
plt.show()

print(ppn.w_)

plt.plot(range(len(ppn.cost_)), ppn.cost_)
plt.xlabel('Iterations')
plt.ylabel('Sum-squared-error')

plt.show()
```

## Stochastic Gradient Descent Perceptron

```python
%matplotlib inline
import matplotlib.pyplot as plt


ppn = Perceptron(epochs=15, eta=0.01, learning='sgd')
ppn.fit(X, y)

plot_decision_regions(X, y, clf=ppn, )
plt.title('Perceptron - Stochastic Gradient Descent')
plt.show()

print(ppn.w_)

plt.plot(range(len(ppn.cost_)), ppn.cost_)
plt.ylabel('Sum-squared-error')

plt.show()
```
